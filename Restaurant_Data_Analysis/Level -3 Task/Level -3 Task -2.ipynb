```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df =pd.read_csv("Dataset .csv")
```

## Task 2: Votes Analysis

Identify the restaurants with the highest and
lowest number of votes.

```python
Highest_voted_restaurant = df.groupby("Restaurant Name")['Votes'].sum().sort_values(ascending=False)
Highest_voted_restaurant = Highest_voted_restaurant.idxmax()
print("The Highest voted Restaurant is :",Highest_voted_restaurant)
```

```python
l_r = df.groupby("Restaurant Name")['Votes'].sum().sort_values()
l_r = l_r.idxmin()
print("The Lowest Voted Restaurant is: ", l_r)
```

Analyze if there is a correlation between the
number of votes and the rating of a
restaurant.

```python
corelation=np.corrcoef(df['Votes'],df['Aggregate rating'])
corelation=np.round(corelation,2)
print(corelation)
```

```python
plt.figure(figsize=(10,5),dpi =100)
sns.heatmap(corelation,annot=True,fmt="2f",linewidth=.4)
plt.show()
```

```python
sns.scatterplot(data=df , y= 'Votes' ,hue='Aggregate rating')
sns.regplot(data=df, x='Aggregate rating', y ='Votes',scatter=False)
plt.title(f"Correlation between Aggregate rating and Votes: 0.31")
plt.show()
print("As per result correlation values = 0.31,which indicate there is a moderate positive correlation between these two variable.")
```
