### Task: Restaurant Chains 
Identify if there are any restaurant chains
present in the dataset.
Analyze the ratings and popularity of
different restaurant chains.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
restaurant_chains = df.groupby("Restaurant Name").size().reset_index(name= "Chain Count")
restaurant_chains = rastaurant_chains [restaurant_chains["Chains Count"]>1]
```

```python
restaurant_chains = restaurant_chains.sort_values(by="Chain Count",ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.bar(restaurant_chains["Restaurant Name"][:10],restaurant_chains["Chain Count"][:10])
plt.xticks(rotation=45,ha="right")
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outelets")
plt.title("Top 10 Restaurant Chains by Number Outlets")
plt.tight_layout()
plt.show()
```

```python
chain_ratings = df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating")
chains_votes = df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes")
```

```python
chain_analysis = pd.merge(chain_ratings, chains_votes, on="Restaurant Name")
```

```python
chain_analysis = chain_analysis.sort_values(by="Average Rating",ascending = False)
```

```python
plt.figure(figsize=(25,9))
plt.bar(chain_analysis["Restaurant Name"][:10],chain_analysis["Average Rating"][:10])
```
