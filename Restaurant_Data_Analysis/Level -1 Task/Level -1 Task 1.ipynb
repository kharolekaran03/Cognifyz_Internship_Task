# Task 1: Top Cuisines
Determine the top three most
common cuisines in the dataset.
Calculate the percentage of
restaurants that serve each of the top
cuisines

```python
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
df = pd.read_csv('Dataset .csv',encoding= 'unicode_escape')
```

```python
cuisine_data = df["Cuisines"]
```

```python
cuisine_counts = cuisine_data.value_counts()
```

```python
top_cuisines = cuisine_counts.head(3)
```

```python
total_restaurent = len(df)
percentage = (top_cuisines / total_restaurent)*100
```

```python
plt.bar(percentage.index,percentage.values)
plt.xlabel("Cuisine")
plt.ylabel("Percentage")
plt.title("Top Three Cuisines")
plt.show           
           
```

```python
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline

df = pd.read_csv('Dataset .csv',encoding= 'unicode_escape')

cuisine_data = df["Cuisines"]

cuisine_counts = cuisine_data.value_counts()

top_cuisines = cuisine_counts.head(3)

total_restaurent = len(df)

percentage = (top_cuisines / total_restaurent)*100

plt.bar(percentage.index,percentage.values)
plt.xlabel("Cuisine")
plt.ylabel("Percentage")
plt.title("Top Three Cuisines")
plt.show           
           
```

```python
plt.pie(top_cuisines, labels = top_cuisines.index, autopct=  "%1.1f%%")
plt.title("Top Three Cuisines")
plt.show()
```
